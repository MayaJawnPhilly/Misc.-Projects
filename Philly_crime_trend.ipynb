```python
import pandas as pd
import os

directory = '/content/drive/My Drive/Philly Crime Data'

all_files = os.listdir(directory)
csv_files = [file for file in all_files if file.endswith('.csv')]

df_list = []
for file in csv_files:
    file_path = os.path.join(directory, file)
    df = pd.read_csv(file_path)
    df_list.append(df)
```

```python
df = pd.concat(df_list, ignore_index=True)
df
```

```python
df['dispatch_date'] = pd.to_datetime(df['dispatch_date'])

df['year'] = df['dispatch_date'].dt.year
df['month'] = df['dispatch_date'].dt.month
df['year_month'] = df['dispatch_date'].dt.strftime('%Y-%m')

df
```

```python
df['text_general_code'].value_counts()
```

```python
import numpy as np

df['firearm'] = np.where((~df['text_general_code'].str.contains('No Firearm'))
                    & (df['text_general_code'].str.contains('Firearm')), 1, 0)

df['violent_crime'] = np.where(
    np.logical_or.reduce([
        df['text_general_code'].str.contains('Aggravated Assault'),
        df['text_general_code'].str.contains('Rape'),
        df['text_general_code'].str.contains('Robbery'),
        df['text_general_code'].str.contains('Homicide')
    ]), 1, 0
)
```

```python
df['property_crime']=np.where(
    np.logical_or.reduce([
        df['text_general_code'].str.contains('Burglary'),
                              df['text_general_code'].str.contains('Larceny'),
                              df['text_general_code'].str.contains('Vehicle|Theft')
                              ]),1,0
)

df
```

```python
df['crime_category'] = np.where(
    np.logical_or.reduce([
        df['text_general_code'].str.contains('Aggravated Assault'),
        df['text_general_code'].str.contains('Rape'),
        df['text_general_code'].str.contains('Robbery'),
        df['text_general_code'].str.contains('Homicide')
    ]), 'violent_crime', df['text_general_code']
)

df['crime_category'] = np.where(
    np.logical_or.reduce([
        df['text_general_code'].str.contains('Burglary'),
        df['text_general_code'].str.contains('Larceny'),
        df['text_general_code'].str.contains('Vehicle|Theft')
    ]), 'property_crime', df['crime_category']
)

df['crime_category'] = np.where(
    np.logical_or.reduce([
        df['text_general_code'].str.contains('DRIVING UNDER')
    ]), 'dui', df['crime_category']
)

df['crime_category'] = np.where(
    np.logical_or.reduce([
        df['text_general_code'].str.contains('Drunk'),
        df['text_general_code'].str.contains('Liquor')
    ]), 'drunk', df['crime_category']
)
```

```python
violent_crime_district_year = df.groupby(['year','dc_dist'])['violent_crime'].sum().reset_index()
violent_crime_district_year
```

```python
#create the rank column
import pandas as pd
grouped_crimes = violent_crime_district_year.groupby(['year', 'dc_dist'])['violent_crime'].sum().reset_index()

sorted_crimes = grouped_crimes.sort_values(['year', 'violent_crime'], ascending=[True, False])
sorted_crimes['crime_rank'] = sorted_crimes.groupby('year')['violent_crime'].rank(method='dense', ascending=False)

violent_crime_district_year = pd.merge(violent_crime_district_year, sorted_crimes[['year', 'dc_dist', 'crime_rank']], on=['year', 'dc_dist'], how='left')
violent_crime_district_year
```

```python
violent_crime_district_year_pivot = pd.pivot_table(violent_crime_district_year, values='crime_rank', index='year', columns='dc_dist', aggfunc='sum')
violent_crime_district_year_pivot
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.heatmap(violent_crime_district_year_pivot,
            annot=True,
            annot_kws={"size": 9},
            cmap='viridis',
            cbar=False,
            fmt=".0f")
plt.title('Violent Crime Rank by District and Year',fontsize=9)
plt.xlabel('District',fontsize=9)
plt.ylabel('Year',fontsize=9)
plt.show()
```
